# Telecom Customer Churn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d=pd.read_csv("E://Datasets//WA_Fn-UseC_-Telco-Customer-Churn.csv")
d.head()
```

```python
d.drop("customerID",axis="columns",inplace=True)
```

```python
d.dtypes
```

```python
pd.to_numeric(d['TotalCharges'],errors='coerce').isnull()
```

```python
d[pd.to_numeric(d['TotalCharges'],errors='coerce').isnull()]
```

```python
d1=d[d['TotalCharges']!=' ']
```

```python
d1.shape
```

```python
d1['TotalCharges'] = pd.to_numeric(d1['TotalCharges'])
```

```python
# d1.replace('No internet service','No',inplace=True)
# d1.replace('No phone service','No',inplace=True)
```

```python
for columns in d1:
    if d1[columns].dtypes=="O":
        print(f'{columns} : {d1[columns].unique()}')
```

```python
yes_no_cols=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_cols:
    d1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
for columns in d1:
    # if d1[columns].dtypes=="O":
    print(f'{columns} : {d1[columns].unique()}')
```

```python
d1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
d2=pd.get_dummies(d1,columns=['InternetService','Contract','PaymentMethod'],dtype=int)
```

```python
# Scaling
col_to_scale=['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()

d2[col_to_scale] = mms.fit_transform(d2[col_to_scale])
```

```python
d2.sample(5)
```

```python
d2.dtypes
```

```python
# Train & Test

x=d2.drop('Churn',axis='columns')
y=d2.Churn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
len(x_train.columns)
```

# Model Building

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])

model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```
